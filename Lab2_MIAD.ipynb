4. У кого в середньому тиск вище, у жінок чи чоловіків.

відрізняються медіанне значення віку курців і тих хто не курить?

```python
gender_group = df.groupby('gender')
men_num = gender_group.height.mean().idxmax()
women_num = gender_group.height.mean().idxmin()
print('Чоловіки відповідають gender #', men_num)
```

```python
overall_count = len(df)
man_count = len(gender_group.get_group(men_num))
print('Скільки чоловіків і жінок в наборі даних?')
print('Чоловіків: ', man_count)
print('Жінок:', overall_count - man_count)
print('Всього:', overall_count)
```

```python
print('Хто в середньому рідше вказує, що вживає алкоголь?')
alco_mean = gender_group['alco'].mean()
difference = abs(alco_mean.loc[men_num] - alco_mean.loc[women_num])
print(f'{"Чоловіки" if alco_mean.idxmin() == men_num else "Жінки"} на {round(difference*100, 4)}%')
```

```python
print('У скільки разів відсоток курців серед чоловіків більше, ніж відсоток курців серед жинок?')
smoke_mean = gender_group['smoke'].mean()
smoking_ratio = smoke_mean.loc[men_num] / smoke_mean.loc[women_num]
print(f'Чоловіки курять у {round(smoking_ratio)} разів частіше за жінок')
```

```python
print('У кого в середньому тиск вище, у жінок чи чоловіків?')
hi_press_mean = gender_group['ap_hi'].mean()
lo_press_mean = gender_group['ap_lo'].mean()
print(f'Систолічний тиск, в середньому, вищий у {"чоловіків" if hi_press_mean.idxmax() == men_num else "жінок"}')
print(f'Діастолічний тиск, в середньому, вищий у {"чоловіків" if lo_press_mean.idxmax() == men_num else "жінок"}')
```

Судячи з чисел, вік вимірюється в днях

```python
smoke_group = df.groupby('smoke')
# 30 днів в місяці, приблизно
non_smoke_mid_months = smoke_group.get_group(0)['age'].transform(lambda a: a / 30)
smoke_mid_months = smoke_group.get_group(1)['age'].transform(lambda a: a / 30)

print(f'Медіанне значення віку курців і тих хто не курить відрізняється на {round(non_smoke_mid_months.median() - smoke_mid_months.median(), 2)} місяців')
```

## Завдання 6
*Цікавлять 2 підгрупи чоловіків, які палять вік від 60 до 64 років включно: перша з верхнім артеріальним тиском строго менше 120 мм рт.ст. і концентрацією холестерину - 4 ммоль/л, а друга - з верхнім артеріальним тиском від 160 (включно) до 180 мм рт.ст. (Не включно) і концентрацією холестерину - 8 ммоль/л. У скільки разів (round) відрізняються частки хворих людей (відповідно до цільової ознаки, cardio) в цих двох підвибірках?*

```python
df['age_years'] = np.round(df['age'] / 365)
elder_men = df.loc[df['age_years'].between(60, 64, inclusive='both')]
low_risk_men = elder_men.loc[(elder_men['ap_hi'] < 120) & (elder_men['cholesterol'] == 1)]
high_risk_men = elder_men.loc[(elder_men['ap_hi'].between(160, 180, inclusive='left')) & (elder_men['cholesterol'] == 3)]
cardio_ratio = round(high_risk_men['cardio'].mean() / low_risk_men['cardio'].mean(), 2)
print(f'Частки здорових людей відрізняються в {cardio_ratio} рази')
```

*Відфільтруйте наступні сегменти пацієнтів (вважаємо помилками в даних):*
* вказане нижнє значення артеріального тиску строго вище верхнього;
* зріст строго менше 2.5% - перцентілі або строго більше 97.5% - перцентілі (використовуйте pd.Series.quantile)
* вага строго менше 2.5% - перцентілі або строго більше 97.5% -
перцентілі

**Скільки відсотків даних (round) було видкінуто?**

```python
filtered_df = df.loc[\
    (df['ap_hi'] >= df['ap_lo']) &\
    (df['height'] >= df['height'].quantile(0.025)) &\
    (df['height'] <= df['height'].quantile(0.975)) &\
    (df['weight'] >= df['weight'].quantile(0.025)) &\
    (df['weight'] <= df['weight'].quantile(0.975))
]
print(f'Під час фільтрування було відкинуто \
{round(100 - (len(filtered_df) / len(df)) * 100, 2)}% даних')
```

## Завдання 9
**Скільки чоловіків і жінок мають зайву вагу?**

```python
pd.options.mode.chained_assignment = None  # default='warn'
def categorize_bmi(bmi):
    if 25 <= bmi < 30:
        return 'Передожиріння'
    elif 30 <= bmi < 35:
        return '1 ступеню'
    elif 35 <= bmi < 40:
        return '2 ступеню'
    elif bmi >= 40:
        return '3 ступеню'
    else:
        return 'Без'

filtered_df['bmi_category'] = filtered_df['BMI'].apply(categorize_bmi)
filtered_df['gender_label'] = filtered_df['gender'].map({ men_num: 'Чоловіки', women_num: 'Жінки' })
obesity_table = pd.crosstab(
    filtered_df['gender_label'],
    filtered_df['bmi_category'],
    margins=False
)

obesity_categories_order = ['Без', 'Передожиріння', '1 ступеню', '2 ступеню', '3 ступеню']
obesity_table = obesity_table.reindex(columns=obesity_categories_order, fill_value=0)
obesity_table = obesity_table.reindex(['Чоловіки', 'Жінки'], fill_value=0)

print("Таблиця розподілу за статтю та категоріями ожиріння:")
print("=" * 80)
print(obesity_table)
```
